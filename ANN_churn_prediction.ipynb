```python
import pandas as pd

df= pd.read_csv("/content/Churn_Modelling.csv")
df.head()
```

```python
## removing unnecessary column
df.drop(columns = ['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
####One hot Encoding
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train_trf.shape
```

```python
######## Defining ANN model

model = Sequential()
#### 3 is defining the perceptron numbers in the input layer..
#### input_dim 11 means that the 3 perceptron will receive 11 input and the activation function is sigmoid here.
model.add(Dense(3,activation='sigmoid',input_dim=11))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
epochs=20
model.fit(X_train_trf,y_train,epochs=20)
```

```python
## to see the wieights of the first layer
model.layers[0].get_weights()
```

```python
## to see the weights of the second layer
model.layers[1].get_weights()
```

```python
### we used sigmoid which gives us the result between 0 to 1. thats why we did not get our output as 0 and 1
model.predict(X_test)
```

```python
### we will use a threshold method to convert the result into 0 and 1
import numpy as np
y_log = model.predict(X_test)
y_pred = np.where(y_log>0.5, 1, 0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python


model = Sequential()
model.add(Dense(11,activation='relu',input_dim=11))
model.add(Dense(15, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss="binary_crossentropy", optimizer="Adam",metrics="accuracy")
```

```python
model.summary()
```

```python
Epochs=200
model.fit(X_train_trf,y_train,epochs=Epochs,validation_split=0.2)
```

```python
model.predict(X_test)
```

```python
### we will use a threshold method to convert the result into 0 and 1
import numpy as np
y_log = model.predict(X_test_trf)
y_pred = np.where(y_log>0.5, 1, 0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
