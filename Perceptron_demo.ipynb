```python
########## Tutorial CampusX
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(data=df, x="cgpa", y="resume_score",hue='placed')
```

```python

X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python

p.fit(X,y)
```

```python
### Weights of the columns.. so we can say CGPA has the highest importance as its weight is high.
p.coef_
```

```python
### It represents the bias value b
p.intercept_
```

```python

from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
