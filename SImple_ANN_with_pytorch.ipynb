```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

```python
X_train_tensor.dtype
```

```python
y_train_tensor.shape
```

# Defining The model


```python
import torch.nn as nn

class MyModel(nn.Module):
    def __init__(self, num_features, output_size):
        super().__init__()
        self.network = nn.Sequential(
        nn.Linear(num_features, 128),
        nn.ReLU(),
        nn.Linear(128, 64),
        nn.ReLU(),
        nn.Linear(64, output_size),
        nn.Sigmoid()
    )
    def forward(self,num_features):
        return self.network(num_features)
```

```python
model= MyModel(X_train_tensor.shape[1],1)
```

```python
from torchinfo import summary

summary(model, input_size=X_train_tensor.shape)
```

```python
import torch.optim as optim
# Define loss function and optimizer
loss = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
# Training loop
num_epochs = 200

for epoch in range(num_epochs):

  y_pred= model.forward(X_train_tensor.float())

  loss_value= loss(y_pred, y_train_tensor.float().view(-1,1))

  optimizer.zero_grad()

  loss_value.backward()

  optimizer.step()

  print(f'Epoch: {epoch + 1}, Loss: {loss_value.item()}')
```

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor.float())
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
